**1. Importing Libraries**

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
```

**2. Importing DataSet and Selecting Features**

```
df=pd.read_csv("titanic.csv")
df['male']=df['Sex']=='male'
feature_names=['Pclass','male','Age']
X=df[feature_names].values
y=df['Survived']
```

**3. Train Test Spliting**

```
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

**4. Using Pre-pruning for visualizing Decision Tree**

```
#Here we use max_depth, min_samples_leaf,max_leaf_nodes
model=DecisionTreeClassifier(max_depth=4,min_samples_leaf=4,max_leaf_nodes=15)
model.fit(X_train,y_train)
fig=plt.figure(figsize=(20,15))
tree.plot_tree(model,feature_names=feature_names)
plt.show()
```
